## BIRCH

**BIRCH** (balanced iterative reducing and clustering using hierarchies) is a clustering algorithm used to perform hierarchical clustering over large data-sets. In this notebook we will experiment with **BIRCH**. 

```python
#imports  
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
#create blobs dataset
from sklearn.datasets import make_blobs
(X,y) = make_blobs(n_samples=10000, 
               n_features=2,
                centers=[[0,0],[5,1],[-3,3],[5,5],[6,-5],[-1,-5],[3,-3]])
```

```python
plt.figure(figsize=(12,8))
plt.title('Initial plot')
sns.scatterplot(X[:,0], X[:,1], hue=y, palette='dark')
```

```python
from sklearn.cluster import Birch
brc = Birch(n_clusters=7)
brc.fit(X)
y_pred = brc.predict(X)
```

```python
plt.figure(figsize=(12,8))
plt.title('Initial plot')
sns.scatterplot(X[:,0], X[:,1], hue=y_pred, palette='dark')
```

```python
#lets see how Birch clustering changes with threshold
thresholds = [0.01,0.1,0.5,1,2]
for t in thresholds:
    brc = Birch(threshold=t, n_clusters=7)
    brc.fit(X)
    y_pred = brc.predict(X)
    plt.figure(figsize=(12,8))
    plt.title(f'Initial plot threshold={t}')
    sns.scatterplot(X[:,0], X[:,1], hue=y_pred, palette='dark')
```

**BIRCH** does a reasonably good job in finding clusters in the data.
