## Tensor Flow 2.0 - Classification
In this notebook we will show a simple classification task done with TF2.

```python
#Do regular imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
print(tf.__version__)
```

Load breast cancer dataset from sklearn

```python
from sklearn.datasets import load_breast_cancer
data = load_breast_cancer()
```

```python
#Perform train test split 
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split( data.data, data.target, test_size=0.33, random_state=42)
N,D = X_train.shape
```

```python
#Scale the data
from sklearn.preprocessing import StandardScaler
scalar = StandardScaler()
X_train_norm = scalar.fit_transform(X_train)
X_test_norm = scalar.transform(X_test)
```

```python
#Setup a Keras model with two layers
# One Input later and one output dense layer with 1 output feature and sigmoid activation
model = tf.keras.models.Sequential([
   tf.keras.layers.Input(shape=(D,)),
   tf.keras.layers.Dense(1, activation='sigmoid')                               
])

#Setup the optimizer as adam, loss as binary_crossentropy and metrics as accuracy
model.compile(optimizer='adam', 
              loss='binary_crossentropy',
              metrics=['accuracy'])

#Fit the model
r = model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=500)
```

```python
#evaluate train and test scores
print(f"Train score {model.evaluate(X_train, y_train)}")
print(f"Test score {model.evaluate(X_test, y_test)}")
```

```python
#Plot the loss as a function of training
plt.plot(r.history['loss'], label='loss')
plt.plot(r.history['val_loss'], label='val-loss')
plt.legend()
```

```python
#Plot accuracy as a function of training
plt.plot(r.history['accuracy'], label='accuracy')
plt.plot(r.history['val_accuracy'], label='val-accuracy')
plt.legend()
```
