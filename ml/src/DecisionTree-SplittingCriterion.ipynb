## Decision Tree splitting Criterion

For **Decision Trees** used for classification there are multiple ways to split the tree (also called impurity criterion). Three of the most common ways are:
1. Entropy
2. Gini impurity
3. Classification Error

In this notebook we will plot the three criterion, to understand them better.

### Entropy 

**Entropy** is defined as:
$\text{Entropy} = -\sum_{i=1}^{c} p(i|t)\text{log}_2p(i|t)$  where $c$ is the number of classes, for two class case the formula reduces to 

$\text{Entropy} = -p\text{log}_2p-(1-p)\text{log}_2(1-p)$  where $p$ is the probability of observing a class 1.

### Gini Impurity

**Gini Impurity** is defined as:
$\text{Gini}=1-\sum_{i=1}^cp(i|t)^2$

For two class case 
$\text{Gini}=1-(p^2+(1-p)^2)$ 


### Classification error
**Classification error** is defined as:
$\text{Error} = 1 - \text{max} {p(i|t)}$ 

Lets plot these metrics for different $p$.

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
p = np.linspace(0.01,0.99,99)
```

```python
#Define entropy
def entropy(p):
    return -p*np.log2(p)-(1-p)*np.log2(1-p)

#Define Gini
def gini(p):
    return (1 - (p**2+(1-p)**2))

#Define classification error
def error(p):
    err = []
    for v in p:
        val = 1 - max(v,1-v)
        err.append(val)
    return np.array(err)
```

```python
plt.figure(figsize=(12,8))
plt.plot(p, entropy(p), label='Entropy')
plt.plot(p, gini(p), label='Gini')
plt.plot(p, error(p), label='Classification Error')
plt.legend()
```

We can scale the entropy plot by 0.5 to see all the three plots on the same figure

```python
plt.figure(figsize=(12,8))
plt.plot(p, 0.5*entropy(p), label='Entropy')
plt.plot(p, gini(p), label='Gini')
plt.plot(p, error(p), label='Classification Error')
plt.legend()
```

We can see from the above plot that **Gini** and **Entropy** are very similar, and these criterion do give similar results while splitting the decision tree.
