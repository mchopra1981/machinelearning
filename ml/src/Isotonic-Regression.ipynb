## Isotonic Regression

**Isotonic Regression** fits a non-decreasing function to data. It solves the following problem.

$\min{w_i*(y_i-\hat{y}_i)^2}$

s.t. $\hat{y_i} \leq \hat{y_j}$ when $X_i \leq X_j$. In this notebook we will try isotonic regression from sklearn.

```python
#import modules
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.isotonic import IsotonicRegression
%matplotlib inline
```

```python
#Create a dataset
np.random.seed(42)
n = 1000
X = np.arange(n)*0.01
y = X + np.random.randn(n)
```

```python
plt.figure(figsize=(12,8))
plt.plot(X,y)
```

```python
#Fit isotonic regression
ir = IsotonicRegression()
ir.fit(X, y)
y_pred = ir.predict(X)

plt.figure(figsize=(12,8))
plt.plot(X,y,label='Data')
plt.plot(X,y_pred,label='Pred')
plt.legend()
```

```python
print(f"Score={ir.score(X,y):.2}")
```
