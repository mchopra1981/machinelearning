## ISOMAP

**ISOMap** is a nonlinear dimensionality reduction method. High Dimensional datasets are very hard to visualize, reducing dimensionality of the datasets to 2-3 dimensions makes visualization easy and easy to gain intuition about the datasets. In this notebook we will experiment with **ISOMAP** algorithm implemented in **SkLearn**.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
from sklearn.manifold import Isomap
from sklearn.datasets import make_s_curve

n_points = 1000
X, color = make_s_curve(n_points, random_state=0)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=color, cmap=plt.cm.Spectral)
ax.view_init(4, -72)
```

```python
isomap = Isomap(n_neighbors=10, n_components=2)
Y = isomap.fit_transform(X)
plt.scatter(Y[:, 0], Y[:, 1], c=color, cmap=plt.cm.Spectral)
```
