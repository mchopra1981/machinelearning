## DBSCAN

**DBSCAN** or **Density-based spatial clustering of applications with noise**. This algorithm doesn't make assumptions about spherical clusters like k-means, not does it partition the dataset into hierarchies that require manual cut-off point. **DBSCAN** algorithm is as follows:

1. A point is considered a **core point** if a specified number (MinPts) of neighboring points fall within the specified radius $\epsilon$.

2. A **border point** is a point that has fewer neighbors than MinPts within $\epsilon$, but lies within the $\epsilon$ radius of a core point.

3. All other points that are neither **core** nor **border** points are considered **noise points**.

After labeling the points as core, border or noise, the **DBSCAN** algorithm is as follows:

1. Form a separate cluster for each core point or connected group of core points.
2. Assign each border point to the cluster of its corresponding core point.

Lets apply **DBSCAN** from sklearn below.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.cluster import DBSCAN

%matplotlib inline
```

```python
n_points_per_cluster = 250

C1 = [-5, -2] + .3 * np.random.randn(n_points_per_cluster, 2)
y1 = np.ones(shape=n_points_per_cluster)
C2 = [4, -1] + .3 * np.random.randn(n_points_per_cluster, 2)
y2 = np.ones(shape=n_points_per_cluster) + 1
C3 = [1, -2] + .3 * np.random.randn(n_points_per_cluster, 2)
y3 = np.ones(shape=n_points_per_cluster) + 2
C4 = [-2, 3] + .3 * np.random.randn(n_points_per_cluster, 2)
y4 = np.ones(shape=n_points_per_cluster) + 3
C5 = [-2, -2] + .3 * np.random.randn(n_points_per_cluster, 2)
y5 = np.ones(shape=n_points_per_cluster) + 4
C6 = [5, 6] + .3 * np.random.randn(n_points_per_cluster, 2)
y6 = np.ones(shape=n_points_per_cluster) + 5
X = np.vstack((C1, C2, C3, C4, C5, C6))
y = np.hstack((y1,y2,y3,y4,y5,y6))
y= y.astype('int')
```

```python
plt.figure(figsize=(12,8))
plt.title('Initial plot')
sns.scatterplot(X[:,0], X[:,1], hue=y, palette='dark')
```

```python
dbscan = DBSCAN(eps=0.3, min_samples=10)
```

```python
y_dbscan = dbscan.fit_predict(X)
```

```python
plt.figure(figsize=(12,8))
sns.scatterplot(X[:,0], X[:,1], hue=y_dbscan, palette='dark')
```

Here all the clusters had same density, and **DBSCAN** identifies them correctly. One limitation of **DBSCAN** is when clusters have different density.
