## Locally Weighted Linear Regression

**Locally Weighted Linear Regression (LOWESS)** 
LOWESS is a non-parametric regression methods that combines multiple linear regression models. LOWESS is one of the best ways to do smoothing, in the notebook below we will apply **LOWESS** regression. 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#Create a test dataset
np.random.seed(1)
sigma = 0.1
X = np.random.uniform(0,1, size=(1000,1))
y = np.sin(2*np.pi*X) + sigma * np.random.randn(1000,1)
plt.figure(figsize=(12,8))
plt.plot(X, y, 'ro')
```

Below we will use StatsModels LOWESS to fit the data.

```python
import statsmodels.api as sm
lowess = sm.nonparametric.lowess
```

```python
fracs = [0.01, 0.03, 0.1, 0.3]
plt.figure(figsize=(12,8))
plt.plot(X, y, 'ro', markersize=1, label='true data')
for f in fracs:
    w = lowess(y.flatten(), X.flatten(), frac=f)
    plt.plot(w[:,0], w[:,1], label=f'Fitted with frac={f}')
plt.legend()
plt.title('LOWESS')
plt.xlabel('x')
plt.ylabel('y')
```

We can see from the above plots that more data we use for each LOWESS kernel. smoother the fit becomes.
