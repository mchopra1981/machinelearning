## Curse of Dimensionality

Say we have a simple algorithm to estimate the value of a function at point $x_\text{test}$, $f(x_\text{test})$  by averaging over all the values which are closer to $x_\text{test}$ in training data. 

$f(x_\text{test}) = \sum_{\mathcal{N}}f(x_\text{train})$, where $\mathcal{N}$ is the neighbourhood of $x_\text{test}$. Problem with this approach is as the dimensionality of $x$ increases the neignbourhood $\mathcal{N}$ is no longer local.   
Below we demonstrate the non-locality of the neignbourhood as dimensionality of $x$ increases.


### How far would you have to go?
How far would you have to go out if you want to sample $v$ of the volume near a point. Lets plot that as a function of dimensionality of $x$. Assume range of $x$ is $(0,1)$, we are trying to answer what is the edge size of the hypercube which will occupy $v$ volume of the hypercube of $x$. 

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

degree = range(1, 100) #Lets look for degress from 1 to 100
volume = [0.01, 0.1, 0.2, 0.3] #Lets look for these fraction of volumes 

e = dict() #Edge of the hyper cube which will have v volume
for v in volume:
    e[v] = list()
    for deg in degree:
        e[v].append(v**(1/deg))

plt.figure(figsize=(20,10))
for v in volume:
    plt.plot(degree, e[v], label=v)
    plt.xlabel('Degree', fontsize=20)
    plt.title('Curse of dimensionality', fontsize=20)
    plt.ylabel('Edge length of hypercube', fontsize=20)
plt.grid()
plt.legend(fontsize=20)
```

We can see from the plot above if dimensions of $x$ is 20, then we would have to look at the neigbourhood of $x$ which is of length 0.8 in each direction to get get 1% volume. Since the range of $x$ is (0,1) and we have to go as far as 0.8 to get 1% volume that is not necessarily local!! And would lead to a bad estimate. This is called curse of dimensionality.
